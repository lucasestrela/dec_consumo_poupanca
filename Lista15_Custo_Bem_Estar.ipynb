# Problema 1) 

Considere a seguinte função de bem-estar:
$$E_0 \sum_{t=0}^{\infty} \beta^t u (C_t)$$

em que $\beta \in (0, 1)$, $u(\cdot)$ é a função utilidade instantânea, $C_t$ é o consumo no período $t$. Considere a sequência de consumo $\{\bar{C}\}_{t=0}^{\infty}$ formada com base no valor esperado do consumo, isto é, $\bar{C}_t = E [C_t]$. Defina o escalar $\lambda$ (positivo) como a quantidade que resolve a seguinte igualdade:

$$E \Bigg[\sum_{t=0}^{\infty} u ((1 + \lambda) C_t) \Bigg] = \sum_{t=0}^{\infty} \beta^{t} u (E [C_t]) \tag{1}$$

Assim, $\lambda$ é o aumento no nível do consumo original, em todos os períodos, que torna o consumidor indiferente entre a nova sequência arriscada, $\{(1 + \lambda) C_t \}_{t=0}^{\infty}$, e a sequência sem risco, $\{\bar{C}\}_{t=0}^{\infty}$ Por esta razão, $\lambda$ foi chamado por Lucas (1987) de custo de bem-estar do ciclo econômico. Para cada cenário abaixo, calcule $\lambda$ usando a condição (1). Assuma hipóteses apropriadas para que os somatórios sejam finitos.

## Item a) 

(Lucas, 1987) Processo estocástico do consumo:
$$C_t = \alpha_0 (1 + \alpha_1)^t \exp (\varepsilon_t)$$ 

em que $\varepsilon_t \sim iidN(0, \sigma^2_{\varepsilon})$

**Caso 1)** Assuma que $u (C_t) = \ln C_t$.

**Caso 2)** Assuma que $u (C_t) = \frac{C_t^{1-\gamma}}{1-\gamma}$, com $\gamma > 0$ e $\gamma \neq 1$.

---

**Resposta**

Começando pelo **Caso 1**

\begin{align*}
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \ln [(1 + \lambda) C_t] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (E [C_t])
\end{align*}

Note que podemos utilizar o processo do consumo para calcular $E [C_t]$, de fato,

\begin{align*}
    E [C_t] & = E[\alpha_0 (1 + \alpha_1)^t \exp (\varepsilon_t)] \\
    & = \alpha_0 (1 + \alpha_1)^t E[\exp (\varepsilon_t)] \\
    & = \alpha_0 (1 + \alpha_1)^t \exp \{ E[\varepsilon_t] + Var(\varepsilon_t)\} \\
    & = \alpha_0 (1 + \alpha_1)^t \exp \{ 1/2 \sigma^2_{\varepsilon} \} \\
\end{align*}

Substituindo o processo gerador do consumo em ambos os lados da igualdade

\begin{align*}
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \bigg[ \ln [1 + \lambda] + \ln[\alpha_0 (1 + \alpha_1)^t \exp (\varepsilon_t)] \bigg]\Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (\alpha_0 (1 + \alpha_1)^t \exp \{ 1/2 \sigma^2_{\varepsilon} \}) \\
      E \Bigg[\sum_{t=0}^{\infty}  \beta^t \bigg[ \ln [1 + \lambda] + \ln \alpha_0 + t \ln[1 + \alpha_1] +\varepsilon_t \bigg] \Bigg] & = \sum_{t=0}^{\infty}\beta^{t} \Bigg[ \ln \alpha_0 + t \ln [1 + \alpha_1] +  1/2 \sigma^2_{\varepsilon} \Bigg] \\
      \sum_{t=0}^{\infty}  \beta^t \bigg[ \ln [1 + \lambda] + \ln \alpha_0 + t \ln[1 + \alpha_1]\bigg] + E \Big[\sum_{t=0}^{\infty}  \beta^t \varepsilon_t \Big]  & = \sum_{t=0}^{\infty} \beta^{t} \Bigg[\ln \alpha_0 + t \ln [1 + \alpha_1] +  1/2 \sigma^2_{\varepsilon} \Bigg] \\
      \sum_{t=0}^{\infty} \beta^t \bigg[ \ln [1 + \lambda] + \ln \alpha_0 + t \ln[1 + \alpha_1]\bigg] & = \sum_{t=0}^{\infty} \beta^{t} \Bigg[ \ln \alpha_0 + t \ln [1 + \alpha_1] + 1/2  \sigma^2_{\varepsilon} \Bigg] \\
      \sum_{t=0}^{\infty} \beta^t \ln [1 + \lambda] & = \sum_{t=0}^{\infty} \beta^{t} \frac{\sigma^2_{\varepsilon}}{2} \\
      \frac{\ln [1 + \lambda]}{1-\beta} & = \frac{\sigma^2_{\varepsilon}}{2(1-\beta)} \\
\end{align*}

Reordenando, obtemos

\begin{align*}
      \ln [1 + \lambda] & = \frac{\sigma^2_{\varepsilon}}{2} \\
      \lambda & = \exp \bigg\{ \frac{\sigma^2_{\varepsilon}}{2} \bigg\} - 1
\end{align*}

Para o **Caso 2**:


\begin{align*}
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \frac{[(1 + \lambda) C_t]^{1-\gamma}}{1-\gamma}\Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \frac{[E(C_t)]^{1-\gamma}}{1-\gamma} \\
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \beta^t E \Big[C_t^{1-\gamma} \Big] & = \sum_{t=0}^{\infty} \beta^{t} [\alpha_0 (1 + \alpha_1)^t \exp \{ 1/2 \sigma^2_{\varepsilon} \} ]^{1-\gamma} \\
\end{align*}

Para calcular $E [C_t^{1-\gamma}]$, 

\begin{align*}
    E [C_t^{1-\gamma}] & = E[\alpha_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp \{\varepsilon_t (1-\gamma)\}] \\
   & = \alpha_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp \bigg \{E[\varepsilon_t (1-\gamma)] + \frac{(1-\gamma)^2}{2} Var(\varepsilon_t ) \bigg \} \\
   & = \alpha_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp \bigg \{\frac{(1-\gamma)^2}{2} \sigma^2_{\varepsilon} \bigg \}
\end{align*}

Substituindo na equação acima

\begin{align*}
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \beta^t \alpha_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp \bigg \{\frac{(1-\gamma)^2}{2} \sigma^2_{\varepsilon} \bigg \} & = \sum_{t=0}^{\infty} \beta^{t} [\alpha_0 (1 + \alpha_1)^t \exp \{ 1/2 \sigma^2_{\varepsilon} \} ]^{1-\gamma} \\
     (1 + \lambda)^{1-\gamma} \alpha_0^{1-\gamma} \exp \bigg \{\frac{(1-\gamma)^2}{2} \sigma^2_{\varepsilon} \bigg \} \sum_{t=0}^{\infty} \beta^t  (1 + \alpha_1)^{t(1-\gamma)}  & = [\alpha_0 \exp \{ 1/2 \sigma^2_{\varepsilon} \}]^{1-\gamma} \sum_{t=0}^{\infty} \beta^{t} [ (1 + \alpha_1)^t ]^{1-\gamma} \\
     (1 + \lambda)^{1-\gamma} & = \exp \bigg\{ \frac{(1-\gamma)}{2} \sigma^2_{\varepsilon} - \frac{(1-\gamma)^2}{2} \sigma^2_{\varepsilon}\bigg\} \\
     (1 + \lambda)^{1-\gamma} & = \exp \bigg\{ \frac{(1-\gamma) - (1-\gamma)^2}{2} \sigma^2_{\varepsilon} \bigg\}  \\
     (1 + \lambda)^{1-\gamma} & = \exp \bigg\{ \frac{(1-\gamma) - [1 - 2 \gamma + \gamma^2]}{2} \sigma^2_{\varepsilon} \bigg\}  \\
     (1 + \lambda)^{1-\gamma} & = \exp \bigg\{ \frac{\gamma (1 - \gamma)}{2} \sigma^2_{\varepsilon} \bigg\}  \\
     \lambda & = \exp \bigg\{ \frac{\gamma}{2} \sigma^2_{\varepsilon} \bigg\} - 1
\end{align*}




## Item b) (Lucas, 1987) 

Processo estocástico do consumo reparametrizado:
$$C_t = \tilde{\alpha}_0 (1 + \alpha_1)^t \exp\{-0.5 \sigma^2_{\varepsilon}\} \exp(\varepsilon_t)$$ 

em que $\varepsilon_t \sim iidN(0, \sigma^2_{\varepsilon})$ e $\tilde{\alpha}_0 \equiv \alpha_0 \exp\{0.5 \sigma^2_{\varepsilon}\}$ 


**Caso 1)** Assuma que $u (C_t) = \ln C_t$.

**Caso 2)** Assuma que $u (C_t) = \frac{C_t^{1-\gamma}}{1-\gamma}$, com $\gamma > 0$ e $\gamma \neq 1$.


---

**Resposta**

Começando pelo **Caso 1**

\begin{align*}
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \ln [(1 + \lambda) C_t] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (E [C_t])
\end{align*}

Note que podemos utilizar o processo do consumo para calcular $E [C_t]$, de fato,

\begin{align*}
    E [C_t] & = E[\tilde{\alpha}_0 (1 + \alpha_1)^t \exp\{-0.5 \sigma^2_{\varepsilon}\} \exp(\varepsilon_t)] \\
    & = \tilde{\alpha}_0 (1 + \alpha_1)^t \exp\{-0.5 \sigma^2_{\varepsilon}\} E[\exp(\varepsilon_t)] \\
    & = \tilde{\alpha}_0 (1 + \alpha_1)^t \exp\{-0.5 \sigma^2_{\varepsilon}\} \exp \bigg\{E[\varepsilon_t] + \frac{1}{2} Var(\varepsilon_t) \bigg\} \\
    & = \tilde{\alpha}_0 (1 + \alpha_1)^t \exp\{-0.5 \sigma^2_{\varepsilon} \} \exp\{ 0.5 \sigma^2_\varepsilon \} \\
    & = \tilde{\alpha}_0 (1 + \alpha_1)^t 
\end{align*}

Substituindo o processo gerador do consumo em ambos os lados da igualdade

\begin{align*}
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \bigg[ \ln [1 + \lambda] + \ln[\tilde{\alpha}_0 (1 + \alpha_1)^t \exp\{-0.5 \sigma^2_{\varepsilon}\} \exp(\varepsilon_t)] \bigg]\Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (\tilde{\alpha}_0 (1 + \alpha_1)^t )  \\
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \bigg[ \ln [1 + \lambda] -0.5 \sigma^2_{\varepsilon} + \varepsilon_t \bigg]\Bigg] & = 0 \\
     \bigg[ \ln [1 + \lambda] -0.5 \sigma^2_{\varepsilon}  \bigg] \sum_{t=0}^{\infty} \beta^t  & = 0 \\
     \frac{\ln [1 + \lambda] -0.5 \sigma^2_{\varepsilon} }{1-\beta} & = 0  \\
\end{align*}

Reordenando, obtemos

\begin{align*}
     \lambda & = \exp \big\{ 0.5 \sigma^2_{\varepsilon} \big\} - 1 \\
\end{align*}

Para o **Caso 2**:


\begin{align*}
      E \Bigg[\sum_{t=0}^{\infty} \beta^t \frac{[(1 + \lambda) C_t]^{1-\gamma}}{1-\gamma}\Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \frac{[E(C_t)]^{1-\gamma}}{1-\gamma} \\
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \beta^t E \Big[C_t^{1-\gamma} \Big] & = \sum_{t=0}^{\infty} \beta^{t} [\tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} ] \\
\end{align*}

Para calcular $E [C_t^{1-\gamma}]$, 

\begin{align*}
    E [C_t^{1-\gamma}] & = E[\tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp\{-0.5 \sigma^2_{\varepsilon} (1-\gamma) \} \exp \{\varepsilon_t (1-\gamma)\} ] \\
    & = \tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp\{-0.5 \sigma^2_{\varepsilon} (1-\gamma) \} \exp \bigg\{E[\varepsilon_t (1-\gamma)] + 0.5 (1-\gamma)^2 \sigma^2_{\varepsilon} \bigg\} \\
    & = \tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp\{-0.5 \sigma^2_{\varepsilon} \gamma(1-\gamma)  \} 
\end{align*}

Substituindo na equação acima

\begin{align*}
  (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \beta^t \Big[\tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} \exp\{-0.5 \sigma^2_{\varepsilon} \gamma(1-\gamma) \} \Big] & = \sum_{t=0}^{\infty} \beta^{t} [\tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} ]  \\
  (1 + \lambda)^{1-\gamma}  \exp \Big\{-0.5 \sigma^2_{\varepsilon} \gamma(1-\gamma) \Big\}  \sum_{t=0}^{\infty} \beta^t \Big[\tilde{\alpha}_0^{1-\gamma}(1 + \alpha_1)^{t(1-\gamma)} \Big] & = \sum_{t=0}^{\infty} \beta^{t} [\tilde{\alpha}_0^{1-\gamma} (1 + \alpha_1)^{t(1-\gamma)} ] \\
  (1 + \lambda)^{1-\gamma} \exp \Big\{-0.5 \sigma^2_{\varepsilon} \gamma(1-\gamma) \Big\} & = 1 \\
  (1 + \lambda) \exp \Big\{-0.5 \sigma^2_{\varepsilon} \gamma \Big\} & = 1\\
  \lambda  & = \exp \Big\{0.5 \sigma^2_{\varepsilon} \gamma \Big\} - 1
\end{align*}


# Problema 2)

$$E_0 \sum_{t=0}^{\infty} \beta^t u(C_t)$$

em que $\beta \in (0, 1)$, $u(\cdot)$ é a função utilidade instantânea, $C_t$ é o consumo no período $t$. Considere a sequência de consumo $\{\bar{C}_t\}_{t=0}^{\infty}$ formada com base no valor esperado do consumo, isto é, $\bar{C}_t = E_0 [C_t]$. $$E_0 \bigg[ \sum_{t=0}^{\infty} \beta^t u((1+\lambda) C_t) \bigg] = \sum^{\infty}_{t=0} \beta^t u(E_0 [C_t])$$

Assim, $\lambda$ é o aumento no nível do consumo original, em todos os períodos, que torna o consumidor indiferente entre a nova sequência arriscada $\{(1+\lambda) C_t\}_{t=0}^{\infty}$, e a sequência sem risco $\{\bar{C}\}_{t=0}^{\infty}$. Assuma que $C_0$ é escolhido antes de $\varepsilon_0$ ser realizado, de modo que $\varepsilon_0$ não pertence ao conjunto de informação do período $t = 0$.

Para cada cenário abaixo, calcule $\lambda$ usando a condição (2). Assuma hipóteses apropriadas para que os somatórios sejam finitos.

## Item a) (Obstfeld, 1994) 

Processo estocástico do consumo:
$$C_t = \alpha_0 (1+\alpha_1)^t \exp \bigg(\sum_{i=0}^{t} \varepsilon_i\bigg)$$

em que $\varepsilon_t \sim iiN(0, \sigma^2_{\varepsilon})$.

**Caso 1)** Assuma que $u (C_t) = \ln C_t$.

**Caso 2)** Assuma que $u (C_t) = \frac{C_t^{1-\gamma}}{1-\gamma}$, com $\gamma > 0$ e $\gamma \neq 1$.

---

Começando pelo **Caso 1**

\begin{align*}
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \ln [(1 + \lambda) C_t] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (E_0 [C_t]) \\
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \Big[\ln (1 + \lambda) + \ln C_t \Big] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (E_0 [C_t]) \\
\end{align*}

Note que podemos utilizar o processo do consumo para calcular $E [C_t]$, de fato,

\begin{align*}
    E_0 [C_t] & = E_0 \bigg[\alpha_0 (1+\alpha_1)^t \exp \bigg(\sum_{i=0}^{t} \varepsilon_i\bigg) \bigg] \\
    & = \alpha_0 (1+\alpha_1)^t E_0 \bigg[\prod_{i}^t \exp \{\varepsilon_i\} \bigg] \\
    & = \alpha_0 (1+\alpha_1)^t \prod_{i}^t E_0 \big[\exp \{\varepsilon_i\} \big] \\
    & = \alpha_0 (1+\alpha_1)^t \prod_{i}^t \exp \{E_0 [\varepsilon_i] + 0.5 \sigma_{\varepsilon, t}^2\} \\
    & = \alpha_0 (1+\alpha_1)^t \exp \{0.5 t \sigma^2_{\varepsilon, t}\} \\
\end{align*}

Substituindo o processo gerador do consumo em ambos os lados da igualdade

\begin{align*}
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \Bigg[\ln (1 + \lambda) + \ln \bigg(\alpha_0 (1+\alpha_1)^t \exp \Big(\sum_{i=0}^{t} \varepsilon_i \Big) \bigg) \Bigg] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln ( \alpha_0 (1+\alpha_1)^t \exp \{0.5 t \sigma^2_{\varepsilon, t}\} ) \\
      \sum_{t=0}^{\infty} \beta^t \Bigg[\ln (1 + \lambda) + \ln \Big[\alpha_0 (1+\alpha_1)^t \Big]  + E_0 \Big(\sum_{i=0}^{t} \varepsilon_i \Big) \Bigg]  & = \sum_{t=0}^{\infty} \beta^{t} \ln ( \alpha_0 (1+\alpha_1)^t \exp \{0.5 t \sigma^2_{\varepsilon, t}\} ) \\
      \sum_{t=0}^{\infty} \beta^t \ln (1 + \lambda) +  \sum_{t=0}^{\infty} \beta^t \ln \Big[\alpha_0 (1+\alpha_1)^t \Big] & = \sum_{t=0}^{\infty} \beta^{t} \Big[\ln [\alpha_0 (1+\alpha_1)^t ] + 0.5 t \sigma^2_{\varepsilon, t} \Big]
\end{align*}




Assumindo homocedasticidade e reordenando, obtemos

\begin{align*}
      \frac{\ln (1 + \lambda)}{1-\beta} & =  \beta \frac{0.5 \sigma^2_{\varepsilon}}{(1-\beta)^2} \\
      \lambda & =  \exp \bigg\{\frac{ \beta }{1-\beta} \frac{\sigma^2_{\varepsilon}}{2} \bigg\} - 1 \\
\end{align*}

Para o **Caso 2**:


\begin{align*}
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \frac{[(1 + \lambda) C_t]^{1-\gamma}}{1-\gamma}\Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \frac{[E_0(C_t)]^{1-\gamma}}{1-\gamma} \\

## Item b) (Obstfeld, 1994) 

Processo estocástico do consumo reparametriza

$$C_t = \tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 (t+1) \sigma^2_{\varepsilon} \} \exp \bigg(\sum^{t}_{i=0} \varepsilon_i \bigg)$$

em que $\varepsilon_t \sim iidN(0, \sigma^2_{\varepsilon})$, $\tilde{\alpha}_0 \equiv \alpha_0 \exp \{0.5 \sigma^2_{\varepsilon}\}$ e $\tilde{\alpha}_1 \equiv (1+\alpha_1) \exp \{0.5 \sigma^2_{\varepsilon}\} - 1$.

**Caso 2)** Assuma que $u (C_t) = \frac{C_t^{1-\gamma}}{1-\gamma}$, com $\gamma > 0$ e $\gamma \neq 1$.

Começando pelo **Caso 1**

\begin{align*}
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \ln [(1 + \lambda) C_t] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (E_0 [C_t]) \\
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \Big[\ln (1 + \lambda) + \ln C_t \Big] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (E_0 [C_t]) \\
\end{align*}

Note que podemos utilizar o processo do consumo para calcular $E [C_t]$, de fato,

\begin{align*}
    E_0 [C_t] & = E_0 \bigg[\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 (t+1) \sigma^2_{\varepsilon} \} \exp \bigg(\sum^{t}_{i=0} \varepsilon_i \bigg)\bigg] \\
    & =\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 (t+1) \sigma^2_{\varepsilon} \}  E_0 \bigg[ \exp \bigg(\sum^{t}_{i=0} \varepsilon_i \bigg)\bigg] \\
    & =\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 (t+1) \sigma^2_{\varepsilon} \}  E_0 \bigg[ \prod^{t}_{i=0} \exp \{\varepsilon_i \}\bigg] \\
    & =\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 (t+1) \sigma^2_{\varepsilon} \} \exp \{0.5 t \sigma^2_{\varepsilon} \} \\
    & =\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{0.5 t \sigma^2_{\varepsilon} -0.5 (t+1) \sigma^2_{\varepsilon} \}  \\
    & =\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{0.5 \sigma^2_{\varepsilon} (t - t- 1) \}  \\
    & =\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 \sigma^2_{\varepsilon} \}  \\
\end{align*}

Substituindo o processo gerador do consumo em ambos os lados da igualdade

\begin{align*}
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \Big[\ln (1 + \lambda) + \ln \Big(\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 (t+1) \sigma^2_{\varepsilon} \} \exp \bigg(\sum^{t}_{i=0} \varepsilon_i \bigg)\Big) \Big] \Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \ln (\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t \exp \{-0.5 \sigma^2_{\varepsilon} \}) \\
      \sum_{t=0}^{\infty} \beta^t \Big[\ln (1 + \lambda) + \ln \Big(\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t\Big) -0.5 (t+1) \sigma^2_{\varepsilon}  + E_0  \bigg(\sum^{t}_{i=0} \varepsilon_i \bigg)\Big) \Big]  & = \sum_{t=0}^{\infty} \beta^{t} [\ln (\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t ) - 0.5 \sigma^2_{\varepsilon}] \\
      \sum_{t=0}^{\infty} \beta^t \Big[\ln (1 + \lambda) + \ln \Big(\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t\Big) -0.5 (t+1) \sigma^2_{\varepsilon} \Big]  & = \sum_{t=0}^{\infty} \beta^{t} \ln (\tilde{\alpha}_0 (1+ \tilde{\alpha}_1)^t ) - \frac{0.5 \sigma^2_{\varepsilon}}{1-\beta} \\
      \sum_{t=0}^{\infty} \beta^t \Big[\ln (1 + \lambda) -0.5 (t+1) \sigma^2_{\varepsilon} \Big]  & = - \frac{0.5 \sigma^2_{\varepsilon}}{1-\beta} \\
     \frac{\ln (1 + \lambda)}{1-\beta} -  \sum_{t=0}^{\infty} \beta^t \Big[0.5 \sigma^2_{\varepsilon}t + 0.5 \sigma^2_{\varepsilon}  \Big]  & = - \frac{0.5 \sigma^2_{\varepsilon}}{1-\beta} \\
     \frac{\ln (1 + \lambda)}{1-\beta} -  \frac{0.5 \sigma^2_{\varepsilon} \beta}{1-\beta^2} + \frac{0.5 \sigma^2_{\varepsilon}}{1-\beta}  & = - \frac{0.5 \sigma^2_{\varepsilon}}{1-\beta} \\
     \ln (1 + \lambda) & = \frac{\beta}{1-\beta} \frac{\sigma^2_{\varepsilon}}{2} \\
     \lambda & = \exp \bigg \{ \frac{\beta}{1-\beta} \frac{\sigma^2_{\varepsilon}}{2} \bigg \} - 1\\
\end{align*}


Para o **Caso 2**:


\begin{align*}
      E_0 \Bigg[\sum_{t=0}^{\infty} \beta^t \frac{[(1 + \lambda) C_t]^{1-\gamma}}{1-\gamma}\Bigg] & = \sum_{t=0}^{\infty} \beta^{t} \frac{[E_0(C_t)]^{1-\gamma}}{1-\gamma} \\
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \beta^t E_0 \Big[C_t^{1-\gamma} \Big] & = \sum_{t=0}^{\infty} \beta^{t} \bigg[\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 (1-\gamma) \sigma^2_{\varepsilon} \}\bigg] \\
\end{align*}

Para calcular $E_0 [C_t^{1-\gamma}]$, 

\begin{align*}
    E_0 [C_t^{1-\gamma}] & = E_0 \bigg[\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 (1-\gamma) (t+1) \sigma^2_{\varepsilon} \} \exp \bigg((1-\gamma)\sum^{t}_{i=0} \varepsilon_i \bigg)\bigg] \\
    & =\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 (1-\gamma) (t+1) \sigma^2_{\varepsilon} \} E_0 \bigg[ \exp \bigg((1-\gamma)\sum^{t}_{i=0} \varepsilon_i \bigg)\bigg] \\
    & =\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 (1-\gamma) (t+1) \sigma^2_{\varepsilon} \}  \exp \bigg\{0.5 t (1-\gamma)^2 \sigma^2_{\varepsilon} \bigg\}  \\
    & =\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{0.5 t (1-\gamma)^2 \sigma^2_{\varepsilon} -0.5 (1-\gamma) (t+1) \sigma^2_{\varepsilon} \} \\
    & =\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{0.5 (1-\gamma) \sigma^2_{\varepsilon} [(1-\gamma) t - (t+1) ] \} \\
    & =\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 (1-\gamma) \sigma^2_{\varepsilon} (\gamma t + 1) \} \\
    & =\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 t  \gamma (1-\gamma) \sigma^2_{\varepsilon} \} \exp \{- 0.5 (1-\gamma) \sigma^2_{\varepsilon} \} 
\end{align*}

Substituindo na equação acima

\begin{align*}
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \beta^t \bigg( \tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 t  \gamma (1-\gamma) \sigma^2_{\varepsilon} \} \exp \{- 0.5 (1-\gamma) \sigma^2_{\varepsilon} \}  \bigg)  & = \sum_{t=0}^{\infty} \beta^{t} \bigg[\tilde{\alpha}_0^{1-\gamma} (1+ \tilde{\alpha}_1)^{t(1-\gamma)} \exp \{-0.5 (1-\gamma) \sigma^2_{\varepsilon} \}\bigg] \\
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} [\beta (1+ \tilde{\alpha}_1)^{1-\gamma}]^{t} \exp \{-0.5 t  \gamma (1-\gamma) \sigma^2_{\varepsilon} \} \exp \{- 0.5 (1-\gamma) \sigma^2_{\varepsilon} \}   & = \sum_{t=0}^{\infty} [\beta (1+ \tilde{\alpha}_1)^{1-\gamma}]^{t} \exp \{-0.5 (1-\gamma) \sigma^2_{\varepsilon} \} \\
     (1 + \lambda)^{1-\gamma}  \sum_{t=0}^{\infty} \Gamma^{t} \exp \{-\kappa \gamma t  \} \exp \{- \kappa \}   & = \sum_{t=0}^{\infty} \Gamma^{t} \exp \{-\kappa \} \\
     (1 + \lambda)^{1-\gamma} \exp \{- \kappa \}  \sum_{t=0}^{\infty} [\Gamma \exp \{-\kappa \gamma \}]^t   & = \frac{ \exp \{-\kappa \}}{1-\Gamma} \\
     (1 + \lambda)^{1-\gamma} \frac{\exp \{- \kappa \} }{1-\Gamma \exp \{-\kappa \gamma \}}   & = \frac{ \exp \{-\kappa \}}{1-\Gamma} \\
     (1 + \lambda)^{1-\gamma} & = \frac{1-\Gamma \exp \{-\kappa \gamma \}}{1-\Gamma}\\
     \lambda & = \Bigg [ \frac{1-\Gamma \exp \{-0.5 \gamma (1-\gamma) \sigma^2_{\varepsilon}  \}}{1-\Gamma} \Bigg]^{\frac{1}{1-\gamma}} - 1
\end{align*}

onde $\Gamma = \beta(1+\alpha_1)^{1-\gamma}$ e $\kappa = 0.5 (1-\gamma) \sigma^2_{\varepsilon}$ e assumindo que $ |\beta \exp \{ \kappa (1-\gamma) t \}| < 1$. 


# Problema 3)

Considere os dois processos estocásticos do 

$$C_t = \tilde{\alpha}_0 (1+\tilde{\alpha}_1)^t \exp \{-0.5 \sigma^2_{\varepsilon}\} \exp \{\varepsilon_t \} \tag{3}$$
$$C_t = \tilde{\alpha}_0 (1+\tilde{\alpha}_1)^t \exp \{-0.5(t+1) \sigma^2_{\varepsilon}\} \exp \bigg(\sum_{i=0}^{t} \varepsilon_i\bigg) \tag{4}$$

## Item a) 

Compare estes dois processos de consumo, identificando as características centrais de cada um deles.

---

**Resposta**
 
Aplicando o logarítmo em (3)
\begin{align*}
      \ln C_t & = \ln \alpha_0 + t \ln (1+\tilde{\alpha}_1) - 0.5 \sigma^2_{\varepsilon} + \varepsilon_t \\
      c_t & = \bar{c} + t \mu - 0.5 \sigma^2_{\varepsilon} + \varepsilon_t,
\end{align*}

onde $c_t \equiv \ln C_t$, $\bar{c} \equiv \ln \alpha_0$ e $\mu \equiv \ln (1+\tilde{\alpha}_1)$. Aplicando o logaritmo em 4 e tomando a diferença
\begin{align*}
  \Delta \ln C_t & = t \ln (1+\tilde{\alpha}_1) - (t-1) \ln (1+\tilde{\alpha}_1) -0.5 (t+1) \sigma^2_{\varepsilon} + 0.5 (t) \sigma^2_{\varepsilon} + \sum_{i=0}^{t} \varepsilon_i - \sum_{i=0}^{t-1} \varepsilon_i \\
  c_t & = c_{t-1} + \mu - 0.5 \sigma^2_{\varepsilon} + \varepsilon_t
\end{align*}

Note que enquanto (3) é estacionário ao redor de sua tendência, (4) possui raiz unitária, portanto, seus choques são cumulativos.


## Item b) 

Em princípio, você diria que um desses processos estocásticos gerariam custos de bem-estar dos ciclos econômicos maiores? Explique sua resposta.

---

**Resposta**

Como (4) possui uma maior persistência do choque, há maior incerteza para o consumidor, e portanto, esperaria que (4) gerasse maiores custos do bem estar.

## Item c) 

Como você investigaria qual destes processos de consumo é mais apropriado para uma data série temporal de consumo?

---

**Resposta**

Dado uma série temporal de consumo, bastaria testar se $\varepsilon_t$ possui raiz unitária por meio de um teste ADF, com tendência. Se mesmo após a inclusão da tendência, o erro continuar não estacionário, haveria evidência a favor de (4).

```python
import numpy as np
import pandas as pd # Para manipular os dados
import matplotlib.pyplot as plt
import statsmodels.api as sm


from statsmodels.tsa.arima_model import ARMA # Versao antiga da fç
# from statsmodels.tsa.deterministic import TimeTrend, DeterministicProcess
```

```python
df_q  = pd.read_csv('https://raw.githubusercontent.com/lucasestrela/dec_consumo_poupanca/main/dados/dataset_trim.csv',
                    delimiter = ",", index_col=0
                    )

# Ajustado o indice como data
df_q.index = pd.to_datetime(df_q.index)
```

```python
N = 50
beta_grid  = np.linspace(0.9, 0.99, N)
gamma_grid = np.linspace(0.5, 20, N)
var_erro_grid = np.linspace(0.0001, 0.01, N)
```

```python
def calcule_custo_lucas(var_erro, gamma):

  if gamma == 1:
    custo = 0.5 * var_erro - 1

  else:
    custo = np.exp(0.5*var_erro*gamma) - 1

  return custo


def calcule_tendencia_linear(data_frame):

  Y = pd.DataFrame([np.log(data_frame)], index = ['consumo']).T
  X = pd.DataFrame([np.ones(len(data_frame))], index = ['constant']).T
  X['trend'] = np.linspace(1, len(data_frame), len(data_frame))

  X.index = Y.index

  model = sm.OLS(Y,X)
  res = model.fit()

  var_erro = res.ssr/(len(df_q['c']) - 2 )

  return var_erro
```

```python
var_erro_c = calcule_tendencia_linear(df_q['c'])
var_erro_k = calcule_tendencia_linear(df_q['k'])
var_erro_s = calcule_tendencia_linear(df_q['s'])

custo1_c = np.zeros((N))
custo1_k = np.zeros((N))
custo1_s = np.zeros((N))

for idx_g, gamma in enumerate(gamma_grid):
  custo1_c[idx_g] = calcule_custo_lucas(var_erro_c, gamma)
  custo1_k[idx_g] = calcule_custo_lucas(var_erro_k, gamma)
  custo1_s[idx_g] = calcule_custo_lucas(var_erro_s, gamma)
```

```python
plt.plot(gamma_grid, custo1_c, label = 'N. Durável')
plt.plot(gamma_grid, custo1_k, label = 'Durável')
plt.plot(gamma_grid, custo1_s, label = 'Serviço')
plt.title('Custo de Bem Estar dos Ciclos Econômicos')
plt.xlabel('$\gamma$')
plt.ylabel('$\lambda$')
plt.legend()
plt.show()
```

```python
fig2 = plt.figure(figsize=(9, 6))

#==================
#  Primeiro subplot
#==================
ax = fig2.add_subplot(1, 1, 1, projection='3d')

X, Y = np.meshgrid(var_erro_grid, gamma_grid)
vf   = np.vectorize(calcule_custo_lucas)
Z    = vf(X, Y)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='magma', edgecolor='none')
ax.set_xlabel('$\sigma_{\\varepsilon}$')
ax.set_ylabel('$\gamma$')
ax.set_zlabel('$\lambda$')
```

# Problema 5) 

Usando dados de consumo da economia brasileira estime os parâmetros do seguinte processo:

$$C_t = \tilde{\alpha}_0 (1+\tilde{\alpha}_1)^t \exp \{-0.5 (t+1)\sigma^2_{\varepsilon}\} \exp \bigg( \sum_{i=0}^{t} \varepsilon_i \bigg)$$

em que $\varepsilon_t \sim iidN (0, \sigma^2_{\varepsilon})$. Considerando um grid para os parâmetros estruturais $\beta$ e $\gamma$, calcule o custo de bem-estar do ciclo econômico, $\lambda$, para a utilidade CRRA (Problema 2). Faça isso para três medidas distintas do consumo: i) consumo total; ii) consumo de não duráveis e serviços; iii) consumo de não duráveis

Custo do Obstfeld

$$\Gamma = \beta(1+\alpha_1)^{1-\gamma}$$

\begin{align*}
\lambda & =
\begin{cases}
\exp \left\{ \frac{1}{2}\frac{\beta }{1-\beta }\sigma _{\varepsilon}^{2}\right\} -1 & \text{ if } \gamma= 1\\
\left[ \frac{1-\Gamma \exp \left\{-0.5\gamma \left( 1-\gamma \right) \sigma _{\varepsilon }^{2}\right\} }{1-\Gamma }\right] ^{\frac{1}{^{1-\gamma }}} -1 & \text{ if } \gamma>0, \gamma \neq 1
\end{cases},
\end{align*}

Da Questão 3) Item a)


\begin{align*}
  \Delta c_t & = \pi_0 + \varepsilon_t,
\end{align*}

onde $\pi_0 \equiv \mu - 0.5 \sigma^2_{\varepsilon}$ e $\mu \equiv \ln(1+\alpha_1)$


```python
def calculate_arima010(data_frame):

    model = ARMA(data_frame, 
                 order = (0,0)
                 )
    
    res = model.fit(transparams = False, 
                    disp=0
                    ) 
    
    var_erro = sum( (data_frame - res.predict())**2 )/(len(data_frame) - 1 )

    # Obtsfeld (1994, pg. 4)
    mu = res.params[0] + 0.5*var_erro

    alpha1 = np.exp(mu) - 1
    
    return var_erro, alpha1


def calcule_custo_obstfeld(var_erro, alpha1, beta, gamma):

  Gamma_grande = beta * (1 + alpha1)**(1-gamma)

  if gamma == 1:
    custo  = np.exp(0.5 * (beta/(1-beta)) * var_erro) - 1

  else:
    numerador   = 1 - Gamma_grande * np.exp(-0.5 * gamma * (1-gamma) * var_erro)
    denominador = 1 - Gamma_grande

    custo = (numerador/denominador)**(1/(1-gamma)) - 1

  return custo
```
