# Problema 1) 

Considere o problema do consumidor:

$$\max_{\{C_t\}_{t=0}^{\infty}} E_0 \sum_{t=0}^{\infty} \beta^t \bigg(-\frac{1}{\theta} e^{-\theta C_t}\bigg)$$

\begin{align*}\text{sujeito a}
    \begin{cases}
        A_{t+1} = (A_t + Y_t - C_t) R_{t+1} \\
        A_0 \text{ dado}
    \end{cases}
\end{align*}

em que $\beta \in (0, 1)$, $\theta > 0$ e a utilidade instantânea é CARA.


## Item a) 

Calcule os coeficientes de aversão absoluta ao risco, aversão relativa ao risco, prudência absoluta e prudência relativa da utilidade CARA. Verifique as condição de Inada.


---

**Resposta**

Seja a função de utilidade CARA definida como:
$$u(c_t) = - \frac{1}{\theta} e^{-\theta c_t},$$

podemos calcular sua derivada parcial de primeira, segunda ordem e terceira:
$$u'(c_t) = e^{-\theta c_t} \qquad u''(c_t) = -\theta e^{-\theta c_t} \qquad u'''(c_t) = \theta^2 e^{-\theta c_t}.$$

Podemos substituir as derivadas parciais nas definições de aversão absoluta ao risco, aversão relativa ao risco, prudência absoluta e prudência relativa.

\begin{align}
     A_A \equiv - \frac{u''(C)}{u'(C)}   & = \frac{\theta e^{-\theta C}}{e^{-\theta C}} = \theta \\
     A_R \equiv - \frac{u''(C)}{u'(C)} C & = \frac{\theta e^{-\theta C}}{e^{-\theta C}} C = \theta C\\
     P_A \equiv - \frac{u'''(C)}{u''(C)} & = \frac{\theta^2 e^{-\theta C}}{\theta e^{-\theta C}} = \theta \\
     P_R \equiv - \frac{u'''(C)}{u''(C)} C & = \frac{\theta^2 e^{-\theta C}}{\theta e^{-\theta C}} C = \theta C\\
\end{align}



## Item b) 

Assumindo solução interior, encontre a equação de Euler do consumidor.

---

**Resposta**

Seja a equação de Bellman:
$$V(A_{t}) = \max_{\{C_t\}_{t=0}^{\infty}} \bigg\{ u(C_t) + \beta E_t[V(A_{t+1})] \bigg\}$$

Calculando a condição de primeira ordem:
\begin{align*}
    0 & = u'(C_t) + \beta E_t\Big[(- R_{t+1}) V' (A_{t+1})  \Big] \\
    u'(C_t) & = \beta E_t\Big[R_{t+1} V' (A_{t+1})  \Big] \tag{1}
\end{align*}

Utilizando o teorema do envelope
$$V'(A_{t}) = \beta E_t[R_{t+1} V'(A_{t+1}) ]  \tag{2}$$

Igualando (1) e (2)
$$V'(A_{t}) = u'(C_t)$$
$$V'(A_{t+1}) = u'(C_{t+1})$$

Substituindo esta última expressão novamente em (1) para obter a equação de Euler
\begin{align*}
    u'(C_t) & = \beta E_t\Big[R_{t+1} u'(C_{t+1})  \Big]
\end{align*}

Como supomos utilidade da forma CARA
\begin{align*}
    e^{-\theta C_t} & = \beta E_t \Big[R_{t+1} e^{-\theta C_{t+1}}  \Big] \\
    1 & = \beta E_t \Big[R_{t+1} e^{-\theta \Delta C_{t+1}}  \Big] \\
\end{align*}



## Item c)* [Caballero, 1990] 

Considere o seguinte guess para a função política do consumo

$$C_t = \Gamma_{t-1} + C_{t-1} + v_t$$

em que $v_t$ é uma inovação em relação ao conjunto de informação $I_{t-1}$. Assuma que a taxa de juros é constante, sendo igual a $\beta^{-1}$, isto é, $R = \beta^{-1}$. Assuma, ainda, que a renda é descrita por um processo MA (1), como segue:

$$Y_t = \bar{Y} + \varepsilon_t + \psi \varepsilon_{t-1}$$

em que $\psi \in (0, 1)$ e $\varepsilon_t \sim iidN(0, \sigma^2_{\varepsilon})$. Considerando este guess e este processo estocástico para a renda, encontre a função consumo. Em outras palavras, determine $\Gamma_{t-1}$ e $v_t$ para todo $t$. Após encontrar a função consumo, calcule e interprete:

$$\frac{\partial C_t}{\partial \sigma^2_{\varepsilon}}$$

Observação: assuma que $\lim_{i\to\infty} R^{-i} A_{t+i} = 0$.

---

**Resposta**

Supondo que $R = \beta^{-1}$ a equação de Euler encontrada no item anterior se torna:
\begin{align*}
    1 & = E_t \Big[e^{-\theta \Delta C_{t+1}}  \Big] \\
\end{align*}

Note que podemos reescrever o *guess* para função política do consumo e substituir na equação de Euler
\begin{align*}
    1 & = E_t \Big[e^{-\theta (\Gamma_{t} + v_{t+1})}  \Big] \\
     & = E_t \Big[e^{-\theta \Gamma_{t}} e^{- \theta  v_{t+1}}  \Big] \\
     & = e^{-\theta \Gamma_{t}}  E_t \Big[e^{- \theta  v_{t+1}}  \Big] \\
\end{align*}

Aplicando o logaritmo em ambos os lados
\begin{align*}
     \theta \Gamma_{t} & =  \ln E_t \Big[e^{- \theta  v_{t+1}}  \Big] \\
      \Gamma_{t} & =  \frac{1}{\theta} \ln E_t \Big[e^{- \theta  v_{t+1}}  \Big] \tag{3}\\
\end{align*}

Suponha que a única fonte de incerteza seja a renda e que esta tenha uma representação de média móvel. Utilizando a restrição orçamentária com a hipótese de juros constantes:
\begin{align*}
\frac{A_{t+1}}{R} & = A_t + Y_t - C_t \\
A_t & = \frac{A_{t+1}}{R} - Y_t + C_t
\end{align*}

Note que podemos reescrever a última equação de forma recursiva
\begin{align*}
A_t & = \frac{1}{R} A_{t+1} - Y_t + C_t \\
& = \frac{1}{R} \bigg[ \frac{1}{R} A_{t+2} - Y_{t+1} + C_{t+1}  \bigg] - Y_t + C_t \\
& = \frac{1}{R^2} A_{t+2} +  \frac{1}{R} (- Y_{t+1} + C_{t+1}) + (- Y_t + C_t) \\
& = \frac{1}{R^2} \bigg[ \frac{1}{R} A_{t+3} - Y_{t+2} + C_{t+2} \bigg] +  \frac{1}{R} (- Y_t + C_t) + (- Y_t + C_t) \\
& = \frac{1}{R^3} A_{t+3} + \frac{1}{R^2} (- Y_{t+2} + C_{t+2}) +  \frac{1}{R} (- Y_t + C_t) + (- Y_t + C_t) \\
& = \quad  \vdots \\
& = \lim_{j \to \infty} \frac{A_{t+j}}{R^j}  + \sum_{i = 0}^{\infty} \frac{1}{R^i} (- Y_{t+i} + C_{t+i})
\end{align*}

Supondo que $\lim_{i \to \infty} \frac{A_{t+i}}{R^{i}} = 0$, podemos escrever a restrição intertemporal como

\begin{align*}
\sum_{i = 0}^{\infty} \frac{1}{R^i} C_{t+i} & = A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} Y_{t+i}
\end{align*}

Note que podemos escrever o processo gerador de dados do consumo de forma recursiva

\begin{align*}
    C_{t+i} & = \Gamma_{t+i-1} + C_{t+i-1} + v_{t+i} \\
    & = \Gamma_{t+i-1} + (\Gamma_{t+i-2} + C_{t+i-2} + v_{t+i-1}) + v_{t+i} \\
    & = \vdots \\
    & = C_{t} + \sum_{j = 0}^{i} \Gamma_{t+i-j} + \sum_{j = 1}^{i} v_{t+j+1} \\
\end{align*}

Substituindo na restrição reescrita

\begin{align*}
\sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg(C_{t} + \sum_{j = 0}^{i} \Gamma_{t+i-j} + \sum_{j = 1}^{i} v_{t+j+1}  \bigg) & = A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} Y_{t+i} \\
\frac{r}{1+r} C_{t} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 0}^{i} \Gamma_{t+i-j} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 1}^{i} v_{t+j+1}  & = A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} Y_{t+i}
\end{align*}

Isolando $C_t$, encontramos a função consumo:

$$C_{t} = \frac{1+r}{r} \bigg[A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg(Y_{t+i} - \sum_{j = 0}^{i} \Gamma_{t+i-j} - \sum_{j = 1}^{i} v_{t+j+1}\bigg)\bigg]$$

Tomando o valor esperado condicionado em $t$
$$C_{t} = \frac{1+r}{r} \bigg[A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg(\bar{Y} - \sum_{j = 0}^{i} E_t[\Gamma_{t+i-j}] - \sum_{j = 1}^{i} E_t[v_{t+j+1}]\bigg)\bigg]$$

Utilizando (3) e substituindo acima
$$C_{t} = \frac{1+r}{r} \bigg[A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg(\bar{Y} - \frac{1}{\theta} \sum_{j = 0}^{i} \ln E_t \Big[e^{- \theta  v_{t+1+i-j}}  \Big] - \sum_{j = 1}^{i} E_t[v_{t+j+1}]\bigg)\bigg]$$

Ajustando o contador
$$C_{t} = \frac{1+r}{r} \bigg[A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg(\bar{Y} -  \sum_{j = 1}^{i} \frac{1}{\theta} \ln E_t \Big[e^{- \theta  v_{t+j+1}}  \Big] - E_t[v_{t+j+1}]\bigg)\bigg]$$
 


Substituindo na restrição reescrita

\begin{align*}
\sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg(C_{t} + \sum_{j = 0}^{i} \Gamma_{t+i-j} + \sum_{j = 1}^{i} v_{t+j+1}  \bigg) & = A_t + \sum_{i = 0}^{\infty} \frac{1}{R^i} Y_{t+i} \\
\frac{r}{1+r} C_{t} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 0}^{i} \Gamma_{t+i-j} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 1}^{i} v_{t+j+1} - \sum_{i = 0}^{\infty} \frac{1}{R^i} Y_{t+i} & = A_t
\end{align*}

Somando e subtraindo $\sum_{i = 0}^{\infty} \frac{1}{R^i} E[Y_{t+i}]$ no LHS da equação:
\begin{align*}
\frac{r}{1+r} C_{t} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 0}^{i} \Gamma_{t+i-j} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 1}^{i} v_{t+j+1} - \sum_{i = 0}^{\infty} \frac{1}{R^i} (Y_{t+i} - E[Y_{t+i}]) - \sum_{i = 0}^{\infty} \frac{1}{R^i} E[Y_{t+i}]& = A_t \\
\frac{r}{1+r} C_{t} + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 0}^{i} \Gamma_{t+i-j} - \sum_{i = 0}^{\infty} \frac{1}{R^i} E[Y_{t+i}] + \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 1}^{i} v_{t+j+1} - \sum_{i = 0}^{\infty} \frac{1}{R^i} (\varepsilon_t + \psi \varepsilon_{t-1}) & = A_t \\
\sum_{i = 0}^{\infty} \frac{1}{R^i} \bigg (C_{t} + \sum_{j = 0}^{i} \Gamma_{t+i-j} -  E[Y_{t+i}] \bigg) + \underbrace{ \sum_{i = 0}^{\infty} \frac{1}{R^i} \sum_{j = 1}^{i} \Big(v_{t+j+1} - (\varepsilon_t + \psi \varepsilon_{t-1}) \Big) }_{=0 \text{ (Caballero, 1990)}}& = A_t \\
\end{align*}

Portanto, deve valer 
$$v_{t+1} = \varepsilon_t + \psi \varepsilon_{t-1}$$

$$f'(\Delta C_{t+1}) = -\theta  \beta  R e^{-\theta \Delta C_{t+1}}$$
$$f''(\Delta C_{t+1}) = \theta^2  \beta R e^{-\theta \Delta C_{t+1}}$$

A expansão de Taylor de segunda ordem para uma função $f(\cdot)$ é dada por

\begin{align*}
      f(\Delta C_{t+1}) & = f(\Delta C_{t+1}^*) + f' (\Delta C_{t+1}^*)(\Delta C_{t+1} - \Delta C_{t+1}^*) + f'' (\Delta C_{t+1}^*)\frac{(\Delta C_{t+1} - \Delta C_{t+1}^*)^2}{2} \\

\begin{align*}

\end{align*}

$$\frac{\partial E [g_{t+1}]}{\partial  Var_t \Big[ \Delta C_{t+1} \Big]} = \frac{\theta}{2} $$

Seja $f(g_{t+1}) = \beta R_{t+1}\big( g_{t+1} + 1\big)^{-\gamma}$. Calculando a derivada parcial de $f(\cdot)$
$$f'(g_{t+1}) = -\gamma \beta R_{t+1}\big( g_{t+1} + 1\big)^{-(\gamma+1)}$$
$$f''(g_{t+1}) = \gamma (\gamma+1) \beta R_{t+1}\big(g_{t+1}  + 1 \big)^{-(\gamma+2)}$$

A expansão de Taylor de segunda ordem para uma função $f(\cdot)$ é dada por

\begin{align*}
      f(g_{t+1}) & = f(g_{t+1}^*) + f' (g_{t+1}^*)(g_{t+1} - g_{t+1}^*) + f'' (g_{t+1}^*) \frac{(g_{t+1} - g_{t+1}^*)^2}{2} \\
      & = \beta R  -\gamma \beta R g_{t+1} +  \gamma (\gamma+1) \beta R \frac{g_{t+1}^2}{2} \\
      & = 1  -\gamma  g_{t+1} +  \gamma (\gamma+1)   \frac{g_{t+1}^2}{2}
\end{align*}

Substituindo na equação de Euler

\begin{align*}
     1 & = E_t [1  -\gamma  g_{t+1} +  \gamma (\gamma+1)   \frac{g_{t+1}^2}{2}] \\
     0 & =  - \gamma  E_t \Big[ g_{t+1}\Big] +  \gamma (\gamma+1) E_t \Big[\frac{g_{t+1}^2}{2}\Big] \\
     \gamma  E_t \Big[ g_{t+1}\Big] & = \gamma (\gamma+1) E_t \Big[\frac{g_{t+1}^2}{2} \Big] \\
     E_t \big[ g_{t+1}\big] & = \frac{\gamma+1}{2}  E_t \big[g_{t+1}^2 \big] \\
     E_t \big[ g_{t+1}\big] & = \frac{\gamma+1}{2}  \Big(Var_t \big[g_{t+1} \big] + E_t^2 \big[g_{t+1} \big]\Big)
\end{align*}

Portanto,

\begin{align*}
     \frac{\partial E_t \big[ g_{t+1}\big]}{\partial Var_t \big[g_{t+1} \big]} & = \frac{\gamma+1}{2}  
\end{align*}

\begin{align*}

\begin{align*}

\end{align*}

 \begin{align*}

onde $\sigma^{C}_{t+1}$ e $\sigma^{r}_{t+1}$ representam as variâncias condicionais em $t$ do $\log$ do consumo e do $\log$ do retorno.
