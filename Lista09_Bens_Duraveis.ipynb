# Problema 1) 

Considere o problema do consumidor

$$\max_{\{D_t\}_{t=0}^{\infty}} E_0 \Bigg\{\sum_{t=0}^{\infty}[ \beta^t u(K_t)] \Bigg\}$$

\begin{align*}
\text{sujeito a}
    \begin{cases}
      A_{t+1} = R(A_t + Y_t - D_t) \\
      K_t = (1-\delta) K_{t-1} + D_t \\
      A_{0}\text{ e } K_{-1} \text{ dados}
    \end{cases}
\end{align*}

em que $K_t$, $A_t$, $Y_t$ e $D_t$ são, respectivamente, o estoque de bens duráveis, a riqueza, a renda e o gasto com bens duráveis no período $t$. O desconto intertemporal $\beta \in (0, 1)$ e a taxa de depreciação $\delta \in [0, 1]$. Finalmente, a função utilidade é tal que $u'(\cdot) > 0$ e $u'' (\cdot) < 0$.




## Item a) 

Faz sentido a função utilidade $u (\cdot)$ depender de $K_t$ ao invés de $D_t$? Explique sua resposta.

---

**Resposta**

Faz sentido, uma vez que o estoque de capital gera um fluxo de serviços utilizado pelo consumidor ao longo do tempo.


## Item b) 

Assumindo solução interior, resolva este problema encontrando a equação de Euler do consumidor. Interprete o resultado obtido.

---

**Resposta**

Escrevendo a equação de Euler

$$V(A_{t}, K_{t-1}) = \max_{\{D_t\}_{t=0}^{\infty}} \{u(K_t) + \beta E_t [V(A_{t+1}, K_{t})]\}$$
\begin{align*}
\text{sujeito a}
    \begin{cases}
      A_{t+1} = R(A_t + Y_t - D_t) \\
      K_t = (1-\delta) K_{t-1} + D_t \\
      A_{0}\text{ e } K_{-1} \text{ dados}
    \end{cases}
\end{align*}

Podemos substituir as restrições de igualdade no problema de maximização e eliminar as restrições.

\begin{align*}
    V(A_{t}, K_{t-1}) = \max_{\{D_t\}_{t=0}^{\infty}} \bigg\{ u \Big[ (1-\delta) K_{t-1} + D_t \Big] + \beta E_t \Big[V  \Big( \big[R (A_t + Y_t - D_t)\big], \big[(1-\delta) K_{t-1} + D_t \big] \Big ) \Big] \bigg\}
\end{align*}

Escrevendo a condição de primeira ordem deste problema:
\begin{align*}
  u'(K_t) + \beta E_t \Big[V_1 (A_{t+1}, K_{t})(-R) + V_2 (A_{t+1}, K_{t})\Big]  = 0 \\
  \Longleftrightarrow \\
  u'(K_t)  +  \beta E_t \Big[V_2 (A_{t+1}, K_{t})\Big]  = \beta R E_t \Big[V_1 (A_{t+1}, K_{t}) \Big] \tag{1}
\end{align*}

Utilizando o Teorema do Envelope para $A_{t}$
\begin{align*}
    V_1 (A_{t}, K_{t-1}) = \beta R E_t \Big[V_1 (A_{t+1}, K_{t}) \Big]  \tag{2}
\end{align*}

e para $K_{t-1}$
\begin{align*}
    V_2 (A_{t}, K_{t-1}) = u'(K_t) (1-\delta) + \beta E_t \Big[V_2(A_{t+1}, K_{t}) \Big]  (1-\delta)\\
  \Longleftrightarrow \\
    V_2 (A_{t}, K_{t-1}) -  u'(K_t) (1-\delta) = + \beta E_t \Big[V_2(A_{t+1}, K_{t}) \Big]  (1-\delta)\\
  \Longleftrightarrow \\
    \frac{V_2 (A_{t}, K_{t-1})}{1-\delta} -  u'(K_t) = \beta E_t \Big[V_2(A_{t+1}, K_{t}) \Big]  \tag{3}
\end{align*}

Note que no RHS de (1) aparece a mesma expressão contida no LHS de (2), e também, no LHS de (1) encontramos a mesma expressão encontrada no RHS de (3).

\begin{align*}
  u'(K_t)  +  \underbrace{\beta E_t \Big[V_2 (A_{t+1}, K_{t})\Big]}_{\frac{V_2 (A_{t}, K_{t-1})}{1-\delta} -  u'(K_t)}  = \underbrace{\beta R E_t \Big[V_1 (A_{t+1}, K_{t})}_{V_1 (A_{t}, K_{t-1})} \Big] \tag{1}
\end{align*}

Assim, podemos encontrar uma relação entre $V_1$ e $V_2$:

\begin{align*}
  V_2 (A_{t}, K_{t-1}) = V_1 (A_{t}, K_{t-1}) (1-\delta)\\
  \Longleftrightarrow \\
  V_2 (A_{t+1}, K_{t}) = V_1 (A_{t+1}, K_{t}) (1-\delta)
\end{align*}

Novamente, podemos utilizar (1) para reescrever em função apenas de $V_1$

\begin{align*}
  u'(K_t)  +  \beta E_t \Big[V_1 (A_{t+1}, K_{t}) (1-\delta)\Big]  = \beta R E_t \Big[V_1 (A_{t+1}, K_{t}) \Big] \\
  \Longleftrightarrow \\
  u'(K_t) = \beta R E_t \Big[V_1 (A_{t+1}, K_{t}) \Big] - \beta E_t \Big[V_1 (A_{t+1}, K_{t}) \Big]  (1-\delta) \\
  \Longleftrightarrow \\
  u'(K_t) = \beta E_t \Big[V_1 (A_{t+1}, K_{t}) \Big] (R - 1 + \delta) \\
  \Longleftrightarrow \\
  u'(K_t) = \beta R E_t \Big[V_1 (A_{t+1}, K_{t}) \Big] \frac{(R - 1 + \delta)}{R} \tag{4}
\end{align*}

e perceba que o RHS de (2) é identico a uma parcela expressao contida acima, portanto, podemos substituir para encontrar
\begin{align*}
  u'(K_t) & = V_1 (A_{t}, K_{t-1}) \frac{(R - 1 + \delta)}{R} \\
  & \Longleftrightarrow \\
   V_1 (A_{t}, K_{t-1})  & = \frac{R}{R - 1 + \delta}  u'(K_t) \\  
  & \Longleftrightarrow \\
   V_1 (A_{t+1}, K_{t})  & = \frac{R}{R - 1 + \delta}  u'(K_{t+1})
\end{align*}

E utilizando a expressao acima em (4), 
\begin{align*}
   u'(K_t) = \beta R E_t \Big[ \underbrace{V_1 (A_{t+1}, K_{t})}_{\frac{R}{R - 1 + \delta}  u'(K_{t+1})} \Big] \frac{(R - 1 + \delta)}{R} \\
   u'(K_t) = \beta R E_t \Big[ \frac{R}{R - 1 + \delta}  u'(K_{t+1}) \Big] \frac{(R - 1 + \delta)}{R}\\
   u'(K_t) = \beta R E_t \Big[  u'(K_{t+1}) \Big]\\
   1  = \beta R E_t \Bigg[ \frac{ u'(K_{t+1})}{u'(K_t)} \Bigg]
\end{align*}

\begin{align*}

$$u'(K_t) = 1 - b K_t$$

\begin{align*}

$$K_{t} = (1-\delta) K_{t-1} + D_{t},$$

$$\Delta D_{t+1} = \Delta K_{t+1} - (1-\delta) \Delta K_t.$$

\end{align*}

$$u'(K_t) = 1 - b K_t$$

\begin{align*}

   1 - b K_t & = \beta R E_t \Big[ 1 - b K_{t+1} \Big]  \\
   1 - b K_t & = \beta R  - \beta R b E_t \Big[  K_{t+1} \Big]  \\
   E_t \Big[  K_{t+1} \Big] & =  \frac{\beta R +  b K_t - 1}{ \beta R b } 
\end{align*}

$$ K_{t+1} = E_t [K_{t+1}] + \varepsilon^K_{t+1}, \forall t$$

# Problema 2) 

Considere o problema do consumidor

$$\max_{\{D_t, C_t\}_{t=0}^{\infty}} E_0 \Bigg\{\sum_{t=0}^{\infty} \beta^t [v(C_t) + u(K_t)] \Bigg\}$$

\begin{align*}\text{sujeito a}
    \begin{cases}
      A_{t+1} = R_{t+1} (A_t + Y_t - C_t - P_t D_t) \\
      K_t = (1-\delta) K_{t-1} + D_t \\
      A_{0}\text{ e }K_{-1}\text{ dados}
    \end{cases}
\end{align*}

em que $K_t$, $A_t$, $Y_t$, $C_t$, $D_t$ e $P_t$ são, respectivamente, o estoque de bens duráveis, a riqueza, a renda, o gasto com bens não-duráveis, o gasto com bens duráveis e o preáo relativo dos bens duráveis no período t. Ainda, $R_{t+1}$ é o retorno $t + 1$ do ativo comprado no período $t$. O desconto intertemporal $\beta \in (0, 1)$ e a taxa de depreciação $\delta \in [0, 1]$. Finalmente, $v'(\cdot) > 0$, $v'' (\cdot) < 0$, $u'(\cdot) > 0$ e $u''(\cdot) < 0$




## Item a) 

Assumindo solução interior, resolva este problema encontrando as equações de Euler do consumidor.

---

**Resposta**

Equação de Bellman:

$$V(A_t, K_{t-1}) = \max_{\{C_t, D_t\}^{\infty}_{t=0}} \Bigg \{ v(C_t) + u(K_t) + \beta E_t [V(A_{t+1}, K_{t})]\Bigg\}$$
\begin{align*}\text{sujeito a}
    \begin{cases}
      A_{t+1} = R_{t+1} (A_t + Y_t - C_t - P_t D_t) \\
      K_t = (1-\delta) K_{t-1} + D_t \\
      A_{0}\text{ e }K_{-1}\text{ dados}
    \end{cases}
\end{align*}

Para eliminar as restrições no problema de maximização da Equação de Bellman, podemos substituí-las na função valor e maximizar em $C_t$ e depois $D_t$

\begin{align*}
    V(A_t, K_{t-1}) = \max_{\{C_t, D_t\}^{\infty}_{t=0}} \Bigg \{ v(C_t) + u \Big( [1-\delta] K_{t-1} + D_t \Big) + \beta E_t \Bigg [ V \bigg( R_{t+1} (A_t + Y_t - C_t - P_t D_t) , [1-\delta] K_{t-1} + D_t  \bigg) \Bigg]\Bigg\}
\end{align*}

A CPO em $C_t$

\begin{align*}
    0 & =  v' (C_t) + \beta E_t \Big[V_1(A_{t+1}, K_{t})(-R_{t+1}) \Big] \\
    v' (C_t) & = \beta E_t \Big[V_1(A_{t+1}, K_{t}) R_{t+1} \Big] \tag{1}
\end{align*}

A CPO em $D_t$

\begin{align*}
    0 & = u'(K_t) + \beta E_t [V_1(A_{t+1}, K_{t})(-P_t R_{t+1}) + V_2(A_{t+1}, K_{t})] \\
    0 & = u'(K_t) - \beta E_t [V_1(A_{t+1}, K_{t})P_t R_{t+1}] + \beta E_t [V_2(A_{t+1}, K_{t})] \\
    \beta E_t [V_1(A_{t+1}, K_{t})P_t R_{t+1}] & = u'(K_t) + \beta E_t [V_2(A_{t+1}, K_{t})] \tag{2}
\end{align*}

Utilizando o Teorema do Envelope e calculando a derivada parcial em $A_{t}$

\begin{align*}
    V_1(A_t, K_{t-1}) = \beta E_t \Bigg [ V_1 (A_{t+1}, K_{t}) R_{t+1}\Bigg] \tag{3}
\end{align*}

 e calculando a derivada parcial em $K_{t-1}$

 \begin{align*}
    V_2(A_t, K_{t-1}) = u'(K_{t}) (1-\delta) + \beta E_t \bigg [ V_2 (A_{t+1}, K_{t}) (1-\delta) \bigg] \\
    V_2(A_t, K_{t-1}) - u'(K_{t})(1-\delta) =   (1-\delta) \beta E_t \bigg [ V_2 (A_{t+1}, K_{t}) \bigg] \\
    \frac{V_2(A_t, K_{t-1})}{1-\delta} - u'(K_{t}) =  \beta E_t \bigg [ V_2 (A_{t+1}, K_{t}) \bigg]  \tag{4}
\end{align*}

Podemos igualar (1) e (3)

\begin{align*}
    v' (C_{t}) & = V_1(A_t, K_{t-1}) \\
    v' (C_{t+1}) & = V_1(A_{t+1}, K_{t})  \tag{5}
\end{align*}

Substituindo a última expressão em (1), obtendo a primeira equação de Euler

\begin{align*}
    v' (C_t) & = \beta E_t \Big[V_1(A_{t+1}, K_{t}) R_{t+1} \Big] \\
    v' (C_t) & = \beta E_t \Big[v' (C_{t+1}) R_{t+1} \Big] \\
    1 & = \beta E_t \Bigg[\frac{v' (C_{t+1})}{v' (C_t)} R_{t+1} \Bigg]  \tag{6}
\end{align*}

Para obter a segunda Equação de Euler, podemos utilizar (5) em (2)

\begin{align*}
    \beta E_t [V_1(A_{t+1}, K_{t})P_t R_{t+1}] & = u'(K_t) + \beta E_t [V_2(A_{t+1}, K_{t})]\\
    \beta E_t [v' (C_{t+1}) P_t R_{t+1}] & = u'(K_t) + \beta E_t [V_2(A_{t+1}, K_{t})]
\end{align*}

Utilizando (4), podemos obter

\begin{align*}
    \beta E_t [v' (C_{t+1}) P_t R_{t+1}] & = u'(K_t) + \Bigg[ \frac{V_2(A_t, K_{t-1})}{1-\delta} - u'(K_{t}) \Bigg] \\
    P_t \beta E_t [v' (C_{t+1}) R_{t+1}] & = \frac{V_2(A_t, K_{t-1})}{1-\delta}  \\
    P_t v' (C_t) & = \frac{V_2(A_t, K_{t-1})}{1-\delta} && \text{[Usando (6)]} \\
    P_t v' (C_t) (1-\delta) & = V_2(A_t, K_{t-1}) \\
    P_{t+1} v' (C_{t+1}) (1-\delta) & = V_2(A_{t+1}, K_{t})
\end{align*}

Substituindo esta última equação em (2)

\begin{align*}
    \beta E_t [V_1(A_{t+1}, K_{t})P_t R_{t+1}] & = u'(K_t) + \beta E_t [V_2(A_{t+1}, K_{t})] \\
    \beta E_t [v' (C_{t+1}) P_t R_{t+1}] & = u'(K_t) + \beta E_t [ P_{t+1} v' (C_{t+1}) (1-\delta)  ]\\
    \beta E_t [v' (C_{t+1}) P_t R_{t+1}] - \beta E_t [ P_{t+1} v' (C_{t+1}) (1-\delta)  ] & = u'(K_t) \\
    \beta E_t \Big[v' (C_{t+1}) (P_t R_{t+1} -  (1-\delta) P_{t+1}) \Big] & = u'(K_t) \tag{7}
\end{align*}

A equação (7) é a segunda Equação de Euler. Note que $P_t R_{t+1}$ é o custo de oportunidade para a compra de um bem durável no período $t$ e $(1-\delta) P_{t+1}$ é o quanto se pode recuperar do bem durável no período seguinte. Portanto $P_t R_{t+1} -  (1-\delta) P_{t+1}$ é uma espécie de custo de oportunidade líquido do bem durável.

## Item b) 

Usando as equação de Euler, derive uma condição intratemporal para a expressão $Q_t = \frac{u'(K_t)}{v'(C_t)}$. Feito isto, explique por que $Q_t$ é chamado custo de uso do serviço de bens duráveis.

---

**Resposta**

Dividndo a equação (7) por $v'(C_t)$

\begin{align*}
    \frac{u'(K_t)}{v'(C_t)} & = \beta E_t \Big[\frac{v' (C_{t+1})}{v'(C_t)} (P_t R_{t+1} -  (1-\delta) P_{t+1}) \Big]  \\

Onde SDF corresponde ao *fator estocástico de desconto*, que é uma taxa que precifica um ativo cujo retorno é $R_{t+1}$. Esta última equação encontrada é o *custo de uso do bem durável*, isto é, o preço pago hoje menos o valor do ativo amanhã (levando em conta a depreciação).

Utilizando a notação do enunciado, podemos escrever 

\begin{align*}
    Q_t = P_t -  (1-\delta) E_t \big[ M_{t+1} P_{t+1} \big] 
\end{align*}

    1 - b_c C_{t} & =  1 - b_c  [C_{t+1} - \varepsilon_{t+1}^{C}] \\

    \Delta C_{t+1} & =  \varepsilon_{t+1}^{C} \\
\end{align*}

Ou seja, o consumo pode ser descrito como um passeio aleatório. Da segunda equação de Euler

\begin{align*}
   u'(K_t) & = \beta E_t \Big[v' (C_{t+1}) (P_t R_{t+1} -  (1-\delta) P_{t+1}) \Big] \\
   u'(K_t) & = \beta [P R -  (1-\delta) P] E_t \Big[v' (C_{t+1})  \Big] \\
   1 - b_k K_t & = P [1 - \beta (1-\delta)]  E_t \Big[(1 - b_c C_{t+1}) \Big]  \\
   1 - P [1 - \beta (1-\delta)] & = b_k K_t - b_c P [1 - \beta (1-\delta)] E_t (C_{t+1}) 
\end{align*}

Então reescrever a esperança igualmente feito acima:
\begin{align*}
   1 - P [1 - \beta (1-\delta)] & = b_k K_t - b_c P [1 - \beta (1-\delta)] [C_{t+1} - \varepsilon_{t+1}^{C}] \\
   1- \mu & = b_k K_t - b_c \mu C_{t+1} + b_c \mu \varepsilon_{t+1}^{C} \\
    b_c \mu C_{t+1} & = \mu - 1+ b_k K_t  +  b_c \mu  \varepsilon_{t+1}^{C} \\
    C_{t+1} & = \frac{\mu - 1}{b_c \mu} + \frac{b_k}{b_c \mu} K_t  + \frac{b_c \mu}{b_c \mu}\varepsilon_{t+1}^C \\
    C_{t+1} & = \frac{\mu - 1}{b_c \mu} + \frac{b_k}{b_c \mu} K_t  + \varepsilon_{t+1}^C,
\end{align*}

onde $\mu =  1 - P [1 - \beta (1-\delta)]$. Note que podemos escrever esta equação acima em diferenças, obtendo
\begin{align*}
    \Delta C_{t+1} & = \frac{b_k}{b_c \mu} \Delta K_t  + \varepsilon_{t+1}^C - \varepsilon_{t}^C,
\end{align*}
e como vimos que o consumo segue um passeio aleatório, temos
\begin{align*}
    \varepsilon_{t+1}^{C} & = \frac{b_k}{b_c \mu} \Delta K_t  + \varepsilon_{t+1}^C - \varepsilon_{t}^C \\
    \Delta K_t & =  \frac{b_c \mu}{b_k}  \varepsilon_{t}^C \\
\end{align*}



    


# Problema 3) 

Considere o problema do consumidor

$$\max_{\{D_t, C_t\}_{t=0}^{\infty}} E_0 \Bigg\{\sum_{t=0}^{\infty} \beta^t u(C_t, K_t) \Bigg\}$$

\begin{align*}\text{sujeito a}
    \begin{cases}
      A_{t+1} = R_{t+1} (A_t + Y_t - C_t - P_t D_t) \\
      K_t = (1-\delta) K_{t-1} + D_t \\
      A_{0}\text{ e }K_{-1}\text{ dados}
    \end{cases}
\end{align*}

em que $K_t$, $A_t$, $Y_t$, $C_t$, $D_t$ e $P_t$ são, respectivamente, o estoque de bens duráveis, a riqueza, a renda, o gasto com bens não-duráveis, o gasto com bens duráveis e o preço relativo dos bens duráveis no período $t$. Ainda, $R_{t+1}$ é o retorno bruto em $t + 1$ do ativo comprado no período $t$. O desconto intertemporal $\beta \in (0, 1)$ e a taxa de depreciação $\delta \in [0, 1]$. Finalmente, a função utilidade é côncava com utilidades marginais de $C_t$ e de $K_t$ positivas.


## Item a) 

Assumindo solução interior, resolva este problema encontrando as equações de Euler do consumidor.

---

**Resposta**

A equação de Bellman deste problema é dada por

$$V(A_{t}, K_{t-1}) = \max_{\{D_t, C_t\}_{t=0}^{\infty}} \bigg\{ u(C_t, K_t) + \beta E_t[ V(A_{t+1}, K_{t}) ]\bigg\}$$

\begin{align*}\text{sujeito a}
    \begin{cases}
      A_{t+1} = R_{t+1} (A_t + Y_t - C_t - P_t D_t) \\
      K_t = (1-\delta) K_{t-1} + D_t \\
      A_{0}\text{ e }K_{-1}\text{ dados}
    \end{cases}
\end{align*}

Note que podemos eliminar as retrições de igualdade substituindo-as diretamente na equação de Bellman

$$V(A_{t}, K_{t-1}) = \max_{\{D_t, C_t\}_{t=0}^{\infty}} \bigg\{ u(C_t, \underbrace{(1-\delta) K_{t-1} + D_t}_{K_t}) + \beta E_t \Big[V( \underbrace{R_{t+1}(A_t + Y_t - C_t - P_t D_t)}_{A_{t+1}}, \underbrace{(1-\delta) K_{t-1} + D_t}_{K_t})\Big] \bigg\}$$

Para maximização deste problema, podemos calcular a condição de primeira ordem em relação a $C_t$

$$0 = u_1 (C_t, K_t) + \beta E_t \Big[V_1 ( A_{t+1}, K_t) (-R_{t+1}) \Big] $$
$$u_1 (C_t, K_t) = \beta E_t \Big[V_1 ( A_{t+1}, K_t) R_{t+1} \Big] \tag{1}$$

E em relação a $D_t$

\begin{align*}
    0 & = u_2(C_t, K_t) + \beta E_t \Big[V_1 ( A_{t+1}, K_t) (- P_t R_{t+1}) + V_2( A_{t+1}, K_t)  \Big] \\
    & = u_2(C_t, K_t) + \beta E_t \Big[V_1 ( A_{t+1}, K_t) (- P_t R_{t+1})\Big] +  \beta E_t \Big[V_2( A_{t+1}, K_t)  \Big]\\
    \beta E_t \Big[V_1 ( A_{t+1}, K_t) (P_t R_{t+1})\Big]  & = u_2(C_t, K_t) +  \beta E_t \Big[V_2( A_{t+1}, K_t)  \Big] \tag{2}
\end{align*}

Utilizando o Teorema de Benveniste Scheinkkman e derivando em relação a $A_t$

$$V_1 (A_{t}, K_{t-1}) = \beta E_t \Big[V( A_{t+1} , K_t) R_{t+1}\Big] \tag{3} $$

Em relação a $K_{t-1}$

\begin{align*}
V_2 (A_{t}, K_{t-1}) & = u_2 (C_t, K_t) (1-\delta) + \beta E_t \Big[V_2( A_{t+1} , K_t) (1-\delta)\Big] \\
\frac{V_2 (A_{t}, K_{t-1})}{1-\delta} - u_2 (C_t, K_t) & =  \beta E_t \Big[V_2( A_{t+1} , K_t) \Big] \tag{4} \\
\end{align*}

É fácil ver que podemos igualar (1) e (3) para encontrar:

$$V_1 (A_{t}, K_{t-1}) = u_1 (C_t, K_t) \Longrightarrow V_1 (A_{t+1}, K_{t}) = u_1 (C_{t+1}, K_{t+1}) \tag{5}$$

E utilizando este resultado em (1) chegamos na primeira Equação de Euler.

\begin{align*}
u_1 (C_t, K_t) & = \beta E_t \Big[V_1 ( A_{t+1}, K_t) R_{t+1} \Big] \\
u_1 (C_t, K_t) & = \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) R_{t+1}\Big]  \tag{6}
\end{align*}

Note que podemos utilizar (5) em (2) para eliminar $V_1$

\begin{align*}
    \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) (P_t R_{t+1})\Big]  & = u_2(C_t, K_t) + \beta E_t \Big[V_2( A_{t+1}, K_t) \Big]
\end{align*}

Utilizando (4), podemos escrever

\begin{align*}
    \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) (P_t R_{t+1})\Big]  & = u_2(C_t, K_t) +  \underbrace{\beta E_t \Big[V_2( A_{t+1}, K_t)}_{\frac{V_2 (A_{t}, K_{t-1})}{1-\delta} - u_2 (C_t, K_t)} \tag{*} \\
    P_t \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) R_{t+1}\Big]  & =  \frac{V_2 (A_{t}, K_{t-1})}{1-\delta} 
\end{align*}

Utilizando a primeira equação de Euler (4), podemos substituir no LHS
\begin{align*}
    u_1(C_t, K_t) P_t  & = \frac{V_2 (A_{t}, K_{t-1})}{1-\delta} \\
    u_1(C_t, K_t) P_t (1-\delta) & = V_2 (A_{t}, K_{t-1})  \\
    u_1(C_{t+1}, K_{t+1}) P_{t+1} (1-\delta) & = V_2 (A_{t+1}, K_{t}) 
\end{align*}

Note que podemos usar este ultimo resultado em (*)

\begin{align*}
    \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) (P_t R_{t+1})\Big]  & = u_2(C_t, K_t) +  \beta E_t \Big[V_2( A_{t+1}, K_t) \Big] \\
    \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) (P_t R_{t+1})\Big]  & = u_2(C_t, K_t) +  \beta E_t \Big[ u_1(C_{t+1}, K_{t+1}) P_{t+1} (1-\delta) \Big] \\

\end{align*}

Perceba que (7) é a segunda equação de Euler


## Item b) 

Usando as equação de Euler, derive uma condição intratemporal para a expressão $Q_t = \frac{u_2(C_t, K_t)}{u_1(C_t, K_t)}$. Feito isto, explique por que $Q_t$ é chamado custo de uso do serviáo de bens duráveis.

\begin{align*}
    u_2(C_t, K_t)  & = \beta E_t \Big[u_1 (C_{t+1}, K_{t+1}) [(P_t R_{t+1}) - P_{t+1} (1-\delta) ]\Big] \\
     \alpha  \Bigg[(1-\alpha) \bigg(\frac{C_{t}}{K_{t}}\bigg)^{\rho} + \alpha  \Bigg]^{\frac{1-\gamma-\rho}{\rho}} K_{t}^{-\gamma}  & = \beta E_t \Bigg[(1-\alpha) \Bigg(1 + \alpha \Bigg[\bigg(\frac{K_{t+1}}{C_{t+1}}\bigg)^{\rho} - 1 \Bigg] \Bigg)^{\frac{1-\gamma - \rho}{\rho}}C_{t+1}^{-\gamma} [(P_t R_{t+1}) - P_{t+1} (1-\delta) ]\Bigg] \\
     \Bigg[(1-\alpha) \bigg(\frac{C_{t}}{K_{t}}\bigg)^{\rho} + \alpha  \Bigg]^{\frac{1-\gamma-\rho}{\rho}} & = \beta E_t \Bigg[\frac{1-\alpha}{\alpha} \Bigg(1 + \alpha \Bigg[\bigg(\frac{K_{t+1}}{C_{t+1}}\bigg)^{\rho} - 1 \Bigg] \Bigg)^{\frac{1-\gamma - \rho}{\rho}} \bigg(\frac{C_{t+1}}{K_{t}}\bigg)^{-\gamma} [(P_t R_{t+1}) - P_{t+1} (1-\delta) ]\Bigg] \\
     0 & = E_t \Bigg[\beta \frac{1-\alpha}{\alpha} \Bigg(\frac{1 + \alpha \Big[\big(K_{t+1}/C_{t+1}\big)^{\rho} - 1 \Big]}{(1-\alpha) \big(C_{t}/K_{t} \big)^{\rho} + \alpha  } \Bigg)^{\frac{1-\gamma - \rho}{\rho}} \bigg(\frac{C_{t+1}}{K_{t}}\bigg)^{-\gamma} [(P_t R_{t+1}) - P_{t+1} (1-\delta) ] - 1\Bigg]  \tag{*}
\end{align*}

Para estimar os parâmetros, podemos utilizar os momentos (*), multiplicá-los por um vetor de instrumentos $\textbf{z}_t$ e estimar por GMM.

---

Para calcular o custo de uso

\begin{align*}
Q_t & = \frac{u_2(C_t,K_t)}{u_1(C_t,K_t)} \\
 & =  \frac{  \alpha[(1-\alpha) C^{\rho} + \alpha K^{\rho}]^{\frac{1-\gamma-\rho}{\rho}}   K^{\rho - 1} }{(1-\alpha) [(1-\alpha) C^{\rho} + \alpha K^{\rho}]^{\frac{1-\gamma - \rho}{\rho}} C^{\rho - 1} } \\
 & =  \frac{  \alpha K^{\rho - 1} }{(1-\alpha) C^{\rho - 1} }
\end{align*}

Note que podemos aplicar o logaritmo dos dois lados

\begin{align*}
Q_t & = \ln \frac{\alpha }{1-\alpha} + (\rho - 1) \ln\frac{K_t}{C_t} \\
& = \ln \frac{\alpha }{1-\alpha} + (\rho - 1) \ln K_t - (\rho - 1) \ln C_t\\
\ln Q_t - \ln P_t  + (\rho - 1) \ln D_t  & = \ln \frac{\alpha }{1-\alpha} + (\rho - 1) \ln K_t - (\rho - 1) \ln C_t + [- \ln P_t + (\rho - 1) \ln D_t ]\\
\underbrace{\ln \bigg(\frac{Q_t}{P_t}\bigg)  + (\rho - 1) \ln \bigg( \frac{D_t}{C_t} \bigg)}_{\text{Estacionario}} & = \ln \frac{\alpha }{1-\alpha}+ (\rho - 1) \ln \bigg( \frac{D_t}{C_t} \bigg)  - \ln P_t \\
v_t & = \ln \frac{\alpha }{1-\alpha}+ (\rho - 1) \ln \bigg( \frac{D_t}{C_t} \bigg)  - \ln P_t
\end{align*}

Como $v_t$ é estacionário, o RHS também é estacionário, portanto, encontramos uma relação de cointegração, e podemos escrever:

\begin{align*}
\ln \bigg( \frac{D_t}{C_t} \bigg) & = - \frac{1}{\rho - 1} \ln \frac{\alpha }{1-\alpha} + \frac{1}{\rho - 1} \ln P_t + \tilde{v_t} \\
 & = \pi_0  + \pi_1 \ln P_t + \tilde{v_t} \\
 & = \pi_0  + \pi_1 \ln \bigg( \frac{P^d_t}{P^c_t} \bigg) + \tilde{v_t}
\end{align*}

Melhor explicando, como o resíduo $\tilde{v_t}$ da equação acima não possui raiz unitária, as variáveis cointegram.

Considere duas frequências: *i)* trimestral; *ii)* anual.

Considere quebras estruturais na relação de cointegração, como feito por Ogaki e Reinhart (1998a, 1998b). Use dummies para os períodos de guerra, ou utilize um método de cointegração que incorpora a possibilidade de ocorrer quebras estruturais no vetor de cointegração.

A regressão a ser rodadada:
\begin{align*}
\ln \bigg( \frac{D_t}{C_t} \bigg) & = \pi_0  + \pi_1 \ln \bigg( \frac{P^d_t}{P^c_t} \bigg) + \tilde{v_t}
\end{align*}

```python
X_q = np.log(df_q['pk']/df_q['pc'])
y_q = np.log(df_q['k']/df_q['c'])

X_a = np.log(df_a['pk']/df_a['pc'])
y_a = np.log(df_a['k']/df_a['c'])

X_q = sm.add_constant(X_q)
X_a = sm.add_constant(X_a)

X_q.columns = ['Const', 'Pd/Pc']
X_a.columns = ['Const', 'Pd/Pc']
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,5))

ax[0].plot(y_q, label='$D_t/C_t$')
ax[0].plot(X_q.iloc[:, 1], label='$P^D_t/P^C_t$')
ax[1].plot(y_a, label='$D_t/C_t$')
ax[1].plot(X_a.iloc[:, 1], label='$P^D_t/P^C_t$')
plt.legend()
plt.show()
```

```python
# Fit and summarize OLS model
mod_q = sm.OLS(y_q, X_q)
mod_a = sm.OLS(y_a, X_a)

res_q = mod_q.fit()
res_a = mod_a.fit()

print('# ------------------------------------- Resultados Trimestrais ------------------------------------- #')
print()
print(res_q.summary())
print()
print()
print('# --------------------------------------- Resultados Anuais --------------------------------------- #')
print()
print()
print(res_a.summary())
```

```python
pi0_q = res_q.params[0]
pi1_q = res_q.params[1]

pi0_a = res_a.params[0]
pi1_a = res_a.params[1]

rho_q = 1 + 1/pi1_q
rho_a = 1 + 1/pi1_a

# Solucao Numerica pra alpha
func_q = lambda alpha : np.log(alpha/(1-alpha)) + pi0_q * (rho_q - 1)
func_a = lambda alpha : np.log(alpha/(1-alpha)) + pi0_a * (rho_a - 1)

alpha_guess = 0.5

alpha_q = fsolve(func_q, alpha_guess)[0]
alpha_a = fsolve(func_a, alpha_guess)[0]

print('# ------------------ Resultados Trimestrais ------------------ #')
print()
print('Rho = ', rho_q)
print('Alpha = ', alpha_q)
print()
print('# --------------------- Resultados Anuais -------------------- #')
print()
print('Rho = ', rho_a)
print('Alpha = ', alpha_a)
```

"The Null hypothesis is that there is no cointegration, the alternative hypothesis is that there is cointegrating relationship. If the pvalue is small, below a critical size, then we can reject the hypothesis that there is no cointegrating relationship."

https://www.statsmodels.org/stable/generated/statsmodels.tsa.stattools.coint.html

```python
# Fazendo o teste de cointegração
eg_test_q = coint(y_q, X_q.iloc[:, 1], trend = 'c')
eg_test_a = coint(y_a, X_a.iloc[:, 1], trend = 'c')

# Não rejeito a hipótese nula de não haver cointegração.
print('# ------------------ Resultados Trimestrais ------------------ #')
print()
print('Estatística do Teste = ', eg_test_q[0])
print('P-Valor              = ', eg_test_q[1])
print('Valor Crítico        = ', eg_test_q[2])
print()
print('# --------------------- Resultados Anuais -------------------- #')
print()
print('Estatística do Teste = ', eg_test_a[0])
print('P-Valor              = ', eg_test_a[1])
print('Valor Crítico        = ', eg_test_a[2])
```
